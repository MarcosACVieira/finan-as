```python
def norm_lb_lf(tab,cap):

    mask = tab.iloc[:, -lb_days:].max(axis=1)-tab.iloc[:, -lb_days:].min(axis=1) > cap

    tab = tab[mask]
    max_lf = tab['lf_val'].quantile(.90).mean()
    min_lf = tab['lf_val'].quantile(.10).mean()
    output = ((tab['lf_val']-min_lf)/(max_lf-min_lf)).clip(0,1).values
    output = tab['lf_val'].values

    max_lb = tab.iloc[:, -lb_days:].quantile(.90).mean()
    min_lb = tab.iloc[:, -lb_days:].quantile(.10).mean()
    input = ((tab.iloc[:, -lb_days:]-min_lb)/(max_lb-min_lb)).clip(0,1).values
    input = tab.iloc[:, -lb_days:].values
    

    return input, output
#############################
lf_gap = 1
lf_days = 20
lb_gap = 2
lb_days = 20

df_ini = df_cap.copy()

# enriquece
df_ini = df_ini[['CNPJ_FUNDO','DENOM_SOCIAL', 'VL_QUOTA','DT_COMPTC']]
df_ini['DT_COMPTC'] = pd.to_datetime(df_ini['DT_COMPTC'])
df_ini['Card'] = df_ini['DT_COMPTC'].rank(ascending=False, method='dense')
df_ini['year'] = df_ini['DT_COMPTC'].dt.year
df_ini.sort_values(by=['CNPJ_FUNDO', 'DT_COMPTC'], inplace=True) 

# input e output
df_ini['lf_val'] = df_ini.groupby('CNPJ_FUNDO')['VL_QUOTA'].shift(-(lf_gap+lf_days))/df_ini.groupby('CNPJ_FUNDO')['VL_QUOTA'].shift(-lf_gap) -1
for i in range(lb_gap,lb_gap+lb_days) :
    df_ini['lb'+str(i)] = df_ini.groupby('CNPJ_FUNDO')['VL_QUOTA'].shift(i)/df_ini.groupby('CNPJ_FUNDO')['VL_QUOTA'].shift(i+1) - 1

```

```python
mask1 = df_ini.iloc[:, -lb_days-1:].isna().sum(axis=1)==0
mask2 = df_ini.iloc[:, -lb_days-1:].eq(np.inf).sum(axis=1)==0
mask3 = df_ini.iloc[:, -lb_days-1:].eq(-np.inf).sum(axis=1)==0
mask4 = df_ini.iloc[:, -lb_days-1:].eq(-1).sum(axis=1)==0
mask5 = df_ini.iloc[:, -lb_days-1:].eq(0).sum(axis=1)<=2
df_clean = df_ini[mask1 & mask2 & mask3 & mask4 & mask5]
df_clean['Card'] = df_clean['DT_COMPTC'].rank(ascending=False, method='dense')
```

```python
model = Sequential()
l_rate = 0.001
thresholds = -50
alphas = 10
from keras import backend as K
model.add(Conv1D(filters=32, kernel_size=6, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds), input_shape=(lb_days, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=64, kernel_size=5, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Flatten())
model.add(Dense(500, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(250, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(100, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(50, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(25, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(10, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dropout(0.3))
model.add(Dense(5, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))
model.add(Dense(1, activation=lambda x: K.relu(x, alpha=alphas, threshold=thresholds)))

model.compile(optimizer = keras.optimizers.Adamax(learning_rate=l_rate), loss = tf.keras.losses.MeanSquaredError())
weights = model.get_weights()
```
